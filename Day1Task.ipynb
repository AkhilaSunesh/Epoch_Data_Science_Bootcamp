## Target Variable: `Score` (Happiness Score)
- Continuous numerical value
- This makes the ML task a **Regression problem**

## Key Features (predictors)
| Feature | Description |
|---|---|
| `GDP per capita` | Economic output per person |
| `Social support` | Having someone to count on in times of trouble |
| `Healthy life expectancy` | Average years of healthy life |
| `Freedom to make life choices` | Satisfaction with freedom |
| `Generosity` | Charitable donations relative to GDP |
| `Perceptions of corruption` | Trust in government & business |

```python
features = ['GDP per capita', 'Social support', 'Healthy life expectancy',
            'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

corr = df19[features + ['Score']].corr()['Score'].drop('Score').sort_values(ascending=False)
print('Correlation with Happiness Score (2019):')
print(corr.round(3))

# Bar chart
fig, ax = plt.subplots(figsize=(9, 4))
colors = ['#2ecc71' if v > 0 else '#e74c3c' for v in corr.values]
corr.plot(kind='bar', ax=ax, color=colors, edgecolor='white', width=0.6)
ax.set_title('Feature Correlation with Happiness Score (2019)', fontsize=13, fontweight='bold')
ax.set_ylabel('Pearson Correlation')
ax.set_xlabel('')
ax.axhline(0, color='black', linewidth=0.8)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('correlation_bar.png', dpi=120, bbox_inches='tight')
plt.show()
print('\nKey insight: GDP, Social Support, and Life Expectancy are the strongest predictors.')
```

## Across years (2017,2018,2019)

```python
fig, ax = plt.subplots(figsize=(9, 4))

for year, df, col in [('2017', df17, 'Happiness.Score'), ('2018', df18, 'Score'), ('2019', df19, 'Score')]:
    ax.hist(df[col], bins=20, alpha=0.55, label=year, edgecolor='white')

ax.set_title('Happiness Score Distribution (2017–2019)', fontsize=13, fontweight='bold')
ax.set_xlabel('Happiness Score')
ax.set_ylabel('Frequency')
ax.legend()
plt.tight_layout()
plt.savefig('score_distribution.png', dpi=120, bbox_inches='tight')
plt.show()
```

## Top 10 vs Bottom 10

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Top 10
top10 = df19.nlargest(10, 'Score')[['Country or region', 'Score']].reset_index(drop=True)
axes[0].barh(top10['Country or region'][::-1], top10['Score'][::-1], color='#27ae60', edgecolor='white')
axes[0].set_title('Top 10 Happiest Countries (2019)', fontweight='bold')
axes[0].set_xlabel('Happiness Score')
axes[0].set_xlim(6.5, 8.0)

# Bottom 10
bot10 = df19.nsmallest(10, 'Score')[['Country or region', 'Score']].reset_index(drop=True)
axes[1].barh(bot10['Country or region'][::-1], bot10['Score'][::-1], color='#e74c3c', edgecolor='white')
axes[1].set_title('Bottom 10 Least Happy Countries (2019)', fontweight='bold')
axes[1].set_xlabel('Happiness Score')
axes[1].set_xlim(2.0, 3.5)

plt.tight_layout()
plt.savefig('top_bottom_countries.png', dpi=120, bbox_inches='tight')
plt.show()
```

## Key Insights on Happiness

1. **Top Drivers** — GDP, health, and social support are the strongest predictors; prosperity alone isn’t enough.  
2. **Persistent Gap** — Average ~5.3, but scores range ~2.7–7.8. Same countries dominate top and bottom, showing structural differences.  
3. **Generosity Weak Link** — Almost no correlation (r ≈ 0.08); cultural/religious norms drive it, making it the least useful predictor.  


## Summary
### Problem Type: **Regression**
The target variable `Score` is continuous. The task is to predict a country's happiness score from its socioeconomic indicators.

**Suitable Algorithms:**
- Linear Regression (strong baseline given high feature correlations)

**Evaluation Metrics:** RMSE, MAE, R²---

### Secondary Problem Type: **Clustering (Unsupervised)**
Without using the score label, we can cluster countries by their feature profiles (GDP, health, social support, etc.) to discover natural happiness "tiers" — useful for policy benchmarking.

**Suggested Algorithm:** K-Means or DBSCAN with PCA for dimensionality reduction.
